# Data cleaning

#drop unnecessary / blank columns
df.drop(['Status','unnamed1'],axis=1,inplace=True)

```python
#check null values
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#change data type
df['Amount']=df['Amount'].astype('int')
```

```python
df.columns
```

```python
df.describe()
```

```python
#use describe for specific columns
df[['Age','Orders', 'Amount']].describe()
```

# Exploratory Data Analysis

GENDER

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
   ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
for bars in ax.containers:
   ax.bar_label(bars)
```

From above graphs it is seen that most of the buyers are females and purchasing power of females are more than men

AGE

```python
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
#Total Amount vs Total Age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

We can conclude grom both the graphs that the maximum spenders are between 26-35 years females

STATE

```python
#Total no of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders' ,ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
#Total amount/sales from top 10 states
sales_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(x='State',y='Amount',data=sales_amount)
```

Above charts depicts taht most of te orders sre from U.P , Maharashtra and Karnataka

MARITAL STATUS

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
sales_amount=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False).head(10)
sns.barplot(x='Marital_Status',y='Amount',data=sales_amount,hue='Gender')
```

Most of the buyers are married women

OCCUPATION

```python
ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales_occupation)
```

Top 3 sectors are IT,Healthcare and Aviation

PRODUCT CATEGORY

```python
ax=sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
sales_category=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount' ,ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_category)
```

Most of the sold products are from Food,Clothing and Electronics Category
